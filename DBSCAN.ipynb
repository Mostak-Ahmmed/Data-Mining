```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Point':['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12'],
    'X': [3,4,5,6,7,6,7,8,3,2,3,2],
    'Y': [7,6,5,4,3,2,2,4,3,6,5,4]
})
```

```python
df
```

```python
from sklearn.cluster import DBSCAN
```

```python
eps_value = 1.9
min_pts = 4
```

```python
clustering = DBSCAN(eps=eps_value, min_samples=min_pts)
```

```python
val = df[['X','Y']]
```

```python
clus = clustering.fit_predict(val)
```

```python
df['cluster'] = clus
```

```python
df
```

```python
df.drop('Point',axis=1,inplace=True)
df.values
```

```python
df_noise = df0 = df[df['cluster']==-1].values
df0 = df[df['cluster']==0].values
df1 = df[df['cluster']==1].values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df0[:,0],df0[:,1],color='red')
plt.scatter(df1[:,0],df1[:,1],color='yellow')
plt.scatter(df_noise[:,0],df_noise[:,1],color='black')
```
