```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'studentid': [101, 102, 103, 104, 105],
    'cgpa': [3.8, 3.5, np.nan, 3.9, 2.7],
    'major': ['CSE', 'EEE', 'business', 'CSe', 'Economics'],
    'internship': [2, 1, 0, 3, np.nan],
    'placed': ['Yes', 'No', 'no', 'Yes', 'No']
}
df = pd.DataFrame(data)
```

```python
df['cgpa'].fillna(df['cgpa'].mean(), inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['cgpa'], color='skyblue', width=0.4, showmeans=True, meanline=True, meanprops={'color': 'red'})
plt.title('Distribution of CGPA', fontsize=16)
plt.xlabel('CGPA', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine()
plt.show()
```
