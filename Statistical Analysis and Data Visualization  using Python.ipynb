```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming updated_df is your DataFrame and it has an 'age' column
# Create a new column 'Categorical_age' based on age ranges
updated_df['Categorical_age'] = pd.cut(updated_df['age'], bins=[0, 18, 30, 50, 100],
                                       labels=['Child', 'Young Adult', 'Adult', 'Senior'])

# Now you can create the catplot
sns.catplot(x="Categorical_age", y="age", kind="box", data=updated_df)

# To display the plot
plt.show()
```
